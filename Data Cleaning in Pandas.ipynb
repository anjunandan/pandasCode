```python
df["Last_Name"]=df["Last_Name"].str.lstrip("...")
df
```

```python
#df["Last_Name"]=df["Last_Name"].str.lstrip("/")
#df["Last_Name"]=df["Last_Name"].str.lstrip("_")

df["Last_Name"]=df["Last_Name"].str.strip("123._/")
df
```

```python
#df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]', '', regex=True)
#df["Phone_Number"].apply(lambda x: x[0:3]+ '-' + x[3:6] + '-' + x[6:10]) 
#df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))

#df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3]+ '-' + x[3:6] + '-' + x[6:10]) 
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')
df
```

```python
#df["Do_Not_Contact"] = df["Do_Not_Contact"].fillna('')
df["Do_Not_Contact"] = df["Do_Not_Contact"].replace("NaN", '', regex=True)
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('Nan','')
df["Phone_Number"] = df["Phone_Number"].str.replace('N/a','')
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('|', '-', regex=False)
```

```python
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace(r'(\d{3})(\d{3})(\d{4})', r'\1-\2-\3', regex=True)
df
```

```python
 df["Last_Name"] = df["Last_Name"].str.replace('/', '', regex=False)
df["Last_Name"] = df["Last_Name"].str.replace('...', '', regex=False)
df["Last_Name"] = df["Last_Name"].str.replace('_', '', regex=False)
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes','Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('No','N')
df
```

```python
df = df.fillna('')
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('N/a','')
df
```

```python
df["Address"] = df["Address"].str.replace('N/a','')
df
```

```python
for x in df.index:
    if df.loc[x,"Do_Not_Contact"] == 'Y':
        df.drop(x, inplace =True)

df
```

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace =True)

df
```

```python
df = df.reset_index(drop = True)
df
```
